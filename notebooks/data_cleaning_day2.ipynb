# Scheme Performace Cleaning

```python
df7 = pd.read_csv("../data/raw/07_scheme_performance.csv")
```

```python
df7.info()
```

```python
df7['return_1yr_pct'] = pd.to_numeric(df7['return_1yr_pct'])
```

```python
df7['return_3yr_pct'] = pd.to_numeric(df7['return_3yr_pct'])
```

```python
df7['return_5yr_pct'] = pd.to_numeric(df7['return_5yr_pct'])
```

```python
df7.info()
```

```python
df7[['return_1yr_pct',
     'return_3yr_pct',
     'return_5yr_pct',
     'benchmark_3yr_pct']].describe()
```

# Observation:
### No anomalous values were found in the return columns.
### All return percentages fall within a reasonable range (approximately 4% to 25%)...

```python
invalid_expense = df7[
    (df7['expense_ratio_pct'] < 0.1) |
    (df7['expense_ratio_pct'] > 2.5)
]

print(invalid_expense)
```

# Observation:
